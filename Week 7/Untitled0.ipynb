```python
# Constant comparison
print('4 == 4 :', 4 == 4) #ตรวจสอบว่าค่าเท่ากันหรือไม่  เท่ากันเป็น True
print('1 < 2:', 1 < 2) # ตรวจสอบค่าแรกน้อยกว่าค่าที่สองหรือไม่ น้อยกว่า true
print('3 > 10:', 3 > 10) #ตรวจสอบค่าแรกมากกว่าค่าที่สองหรือไม่ มากกว่าเป็น True
print('2 <= 1.5', 2 <= 1.5) #ตรวจสอบเช็คค่าแรกน้อยกว่าหรือเท่ากับค่าที่สองหรือไม่ น้อยกว่าหรือเท่ากับ true
print()

# Variable comparison
a = 10
b = 8
print('a != b:', a != b)  # ตรวจสอบว่าค่าไม่เท่ากันหรือไม่ ไม่เท่า เป็น  true
print('a - b == 2:', a - b == 2)
print()
```

```python
class Person: # สร้างฟังชั่นชื่อ person
    __age = 4
    def __init__(self, name): #ฟังชั่นเริ่มต้น
      if name == 'Na':
        self.name = 'no name'
      else:
        self.name = name
      
    def setName(self, name):# ฟังชั่น ตรวจสอบ name เป็น na หรือไม่ ถ้าใช่จะเปลี่ยนเป็น no name
      if name == 'Na':
        self.name = 'no name'
      else:
        self.name = name

    def getName(self): # ฟังชั่นรับค่า
        return self.name 

    def setAge(self, age):# ฟังชั่นตรวจสอบอายุไม่ให้เกิน 120 ปี 
      if age <= 120:
        self.__age = age
    
    def getAge(self):
      return self.__age


p1 = Person('Tommy') # สร้างตัวแปรclass ชื่อ P1 เป็น tommy
print(p1.getName()) #เรียกใช้ฟังชั้น
p2 = Person('Na')
print(p2.getName())
p2 = Person('Dum')
print(p2.getName())

#print(p2.name)
#p2.name = 'xxx'
#print(p2.a)

print(p2.getAge())
p2.setAge(115)
print(p2.getAge())
```

```python
n = int(input("Enter a number: "))# รอรับค่าจากuserเป็นจำนวนเต็ม

if n % 2 == 0:
    print('%d is even number' % n) 
else:
    print('%d is odd number' % n)

if n > 0:
    print('%d is positive number' % n)
elif n == 0:
    print('%d is zero' % n)
else:
    print('%d is negative number' % n)
```

Bitwise operators

```python
a = 3 # 00000011
b = 5 # 00000101

print('a & b =', a & b)
print('a | b =', a | b)
print('a ^ b =', a ^ b)
print('~a =', ~a)#แปลงจำนวนติดลบ
print('a << 1 =', a << 1) # เลื่อนบิทไปทางขวา1ตำแหน่ง
print('a << 2 =', a << 2) # เลื่อนบิทไปทางขวา2ตำแหน่ง
print('100 >> 1 =', 100 >> 1) #เลื่อนบิทตำแหน่งที่100ไปทางซ้าย 1 ตำแหน่ง
```

Sequence Operators

```python
names = ['Mateo', 'David', 'Andrill', 'Joshep']

if 'Mateo' in names:
    print('\'Mateo\' exist in the list')
else:
    print('\'Mateo\' not exist in the list')

if 'Jonathan' in names:
    print('\'Jonathan\' exist in the list')
else:
    print('\'Jonathan\' not exist in the list')

numbers = {'1': 'one', '3': 'three', '2': 'two', '5': 'five'}

if 'one' in numbers.values(): #ตรวจสอบค่าในตัวแปร one เป็นตัวเลขหรือไม่
    print('\'one\' exist in the dictionary values')
else:
    print('\'one\' not exist in the dictionary values')

if '7' in numbers.keys(): #ตรวจสอบค่าในตัวแปร one เป็นตัวเลขหรือไม่
    print('\'7\' exist in the dictionary keys')
else:
    print('\'7\' not exist in the dictionary keys')
```
